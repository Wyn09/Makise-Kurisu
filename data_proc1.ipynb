# 删除音频时长小于5秒的

```python
import glob
import os
import librosa 
```

```python
# 获取所有音频文件路径
file_paths = glob.glob("./voice/*")

# 遍历每个文件
for file_path in file_paths:
    # 加载音频文件
    audio, sr = librosa.load(file_path, sr=None)
    
    # 计算音频时长（秒）
    duration = librosa.get_duration(y=audio, sr=sr)
    
    # 如果时长小于5秒，则删除文件
    if duration < 5:
        os.remove(file_path)
        print(f"Deleted {file_path} (Duration: {duration:.2f} seconds)")
    else:
        print(f"Kept {file_path} (Duration: {duration:.2f} seconds)")
```
