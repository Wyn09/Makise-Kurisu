## 将解包出的文本翻译为中文


```python
from openai import OpenAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv("./.env")
```

```python
client = OpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url="https://api.deepseek.com"
)
def simple_chat_completion(client, model_name, message):
    res = ""

    for r in client.chat.completions.create(
        model = model_name,
        messages=[
            {
                "role":"user",
                "content":message
            }
        ],
        stream=True
    ):
        if r.choices[0].delta.content is not None:
            res += r.choices[0].delta.content
            print(r.choices[0].delta.content, end="")
    
    return res
```

```python
simple_chat_completion(client, "deepseek-reasoner", "你是谁？")
```

```python
def chat_write_completion(client, model_name, messages):
    res = ""
    for r in client.chat.completions.create(
        model = model_name,
        messages=messages,
        stream=True
    ):
        out = r.choices[0].delta.content
        if out is not None:
            res += out
            with open("./data/translated.txt", "a", encoding="utf-8") as f:
                f.write(out)
            print(out, end="")
    return res
    
    
```

```python
import time
def trans(window_size = 100):

    translated_text = ""

    system_prompt = """
        你是一个翻译助手，你能将文本翻译地信达雅，并且贴合情景。
            """
   
    for i in range(13300, len(data), window_size):
        time.sleep(3)
        print(f"\ni: {i}")
        text = data[i: i + window_size]
        messages = [
            {
                "role":"system",
                "content":system_prompt

            },
            {
                "role":"user",
                "content":f"""以下是命运石之门的英文小说，保持格式一致，翻译成中文。角色名使用全名，如牧濑红莉栖，冈部伦太郎，桥田至。举例：
                
                [name]椎名真由理[line]“那么，原本拥有普通记忆的真由氏会被叠加了额外记忆的真由氏覆盖吗？”[%p]
                [name]冈部伦太郎[line]“没错。被抹消了。换言之你将重历这一周。可以重复同样的事，也可以做完全不同的事。”[%p]
                [name]冈部伦太郎[line]“并且保留关于被抹消那周的记忆。”[%p]
                        
                删掉color index。只给我翻译后的结果，不要添加其他说明。{text}"""
                
            }
        ]
        translated_text += chat_write_completion(client,  "deepseek-reasoner", messages)
    return translated_text
```

```python
translated_text = trans(100)
```
